## Python Environments

It's good practice to use python virtual environments to install the packages you need to import into your projects. This saves installing every python package onto your base computer, and sometimes specific package versions might be needed, so creating virtual environments is a good way to manage your python packages.

### Troubleshooting

First, try resetting Navigator.

$ anaconda-navigator --reset

The above command should remove corrupted log files at ~/.anaconda, but if this does not work or results in an error, try removing the log files manually.

$ rm -rf /home/$USER/.anaconda

Finally if that doesn't work try:

$ conda install anaconda
